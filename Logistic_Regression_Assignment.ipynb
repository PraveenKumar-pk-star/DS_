```python
# 2. Data Preprocessing

# a. Handle missing values (e.g., imputation)

df_train["Age"].fillna(df_train["Age"].mean(),inplace=True)
df_train["Embarked"].fillna(df_train["Embarked"].mode()[0],inplace=True)
df_train.drop(columns=["Cabin"],inplace=True)
df_train.isnull().sum()
```

```python
df_train.tail()
```

```python
# b. Encode categorical variables

df_train = pd.get_dummies(df_train, columns=['Sex', 'Embarked'], drop_first=True)
print(df_train.head())
```

```python
# 3. Model Building

# a. Build a logistic regression model using appropriate libraries (e.g., scikit-learn)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score,accuracy_score
X=df_train.drop(columns=['PassengerId', 'Name', 'Ticket', 'Survived'])
Y=df_train["Survived"]
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
model=LogisticRegression()
model.fit(x_train,y_train)
y_predit=model.predict(x_test)
print(r2_score(y_test,y_predit))
print(x_train)
```

```python
# 4. Model Evaluation

# a. Evaluate the performance of the model
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score, roc_auc_score, roc_curve
y_predit=model.predict(x_test)
y_prob=model.predict_proba(x_test)[:,1]
accuarcy=accuracy_score(y_test,y_predit)
precision=precision_score(y_test,y_predit)
recall=recall_score(y_test,y_predit)
f1=f1_score(y_test,y_predit)
roc=roc_auc_score(y_test,y_predit)
print(f"accuarcy:{accuarcy:2f}")
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-score: {f1:.2f}')
print(f'ROC-AUC score: {roc:.2f}')
tpr,fpr,thresholds=roc_curve(y_test,y_prob)
plt.figure(figsize=(10, 6))
plt.plot(tpr,fpr,label="logistic regression")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()



```

```python

# 5. Interpretation

# a. Interpret the coefficients of the logistic regression model

def_cofe=pd.DataFrame({"features":X.columns,
                       "cofficent":model.coef_[0]})
print(def_cofe)
```
